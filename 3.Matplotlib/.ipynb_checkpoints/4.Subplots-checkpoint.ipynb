```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#subplot产生多子图
x = np.arange(0,2,0.01)
y1 = x**2 + 1.0
y2 = x**3 + 1.0

#先subplot 再画图
plt.subplot(121)
plt.plot(x,y1)

plt.subplot(122)
plt.plot(x,y2)
```

```python
# random, seed 使得多次试验的随机值在同一位置上都相同
np.random.seed(18)
x = np.random.rand()
x
```

```python
dt = 0.001
t = np.arange(0,1,dt)
h = np.exp(-t/0.05)
s1 = np.random.randn(len(t))
y = np.convolve(s1,h)[:len(t)] *dt #有色噪声

plt.axis([0,1, 1.1 *np.min(y) , 2* np.max(y)] )
plt.plot(t,y)

#axes是做图中图的命令，画一下y的直方图
plt.axes([.65,.6,.2,.2],facecolor = 'k')
n,bins,patches =plt.hist(y,50,normed = 1)

#画一下h的时序图
plt.axes([.2,.6,0.2,.2],facecolor = 'k')
plt.plot(t,h)
plt.xlim(0,0.2) #限制x的范围
plt.ylim(0,2) #限制y的范围
plt.show()
```
