```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## HistGram

# Histgram是经常用来绘制pdf（概率密度函数）的

```python
c = np.random.seed(2018)
(mu,sigma) = (100,2)
data = mu + sigma * np.random.randn(10000)
```

```python
# the histgram of the data
#n是每个bin里出现的数值的概率
#bins 是 data最小值到最大值之间分的区间的个数
#100个区间，就有101个分割值，
#返回值里边的bins就有包括从最小值到最大值的101个值
#patches

(n, bins, patches) = plt.hist(data,bins = 100,normed=True,facecolor = 'b',alpha = 0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Hist')
plt.grid(True)
plt.show()
```

```python
print ('n = ',n)
```

```python
print('bin = ',bins)
bins.size
data.max()
```

```python
#对天气数据统计一下概率直方图
weatherdata = pd.read_csv('weather_2012.csv')
(n,bins,patches) = plt.hist(weatherdata['Temp (C)'],bins = 50,facecolor = 'g',alpha = 0.5)

plt.xlabel('Temparature in C')
plt.ylabel('Counts of days')
plt.title('Hist')
plt.setp(patches[5],facecolor = 'r',alpha = 0.5)
plt.grid(True)
plt.show()
```
