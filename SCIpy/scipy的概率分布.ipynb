```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#导入Scipy的统计模块
#statistics
import scipy.stats as st
```

```python
#生成自变量范围和对应的pdf
x = np.arange(-3,3,0.01)
x_pdf = st.norm.pdf(x)
y  = np.arange(-2,2,0.01)
d = st.norm.pdf(y)
```

```python
plt.figure()
plt.plot(x,x_pdf)
plt.figure()
plt.plot(y,d)
```

```python
x_norm = st.norm.rvs(size = 500)
h = plt.hist(x_norm,normed = True,bins = 20)
plt.plot(x,x_pdf)
```

```python
#在这组数据下,正态分布的最大似然估计值是
x_mean, x_std = st.norm.fit(x_norm)
print ('mean, ', x_mean)
print ('x_std, ', x_std)
```
