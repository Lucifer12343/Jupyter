```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 100)
y = 4 * x + 1.5
noise_y = y + np.random.randn(y.shape[-1]) * 2.5
```

```python
plt.scatter(x,noise_y,s = 10 ,c = 'k')
```

```python
coefficiency = np.polyfit(x,noise_y,1)
```

```python
coefficiency
```

```python
p = plt.plot(x, noise_y, 'rx')
p = plt.plot(x, coefficiency[0] * x + coefficiency[1], 'k-')
p = plt.plot(x, y, 'b--')
```

```python
#使用
np.poly1d?
f = np.poly1d(coefficiency)

p = plt.plot(x, noise_y, 'rx')
plt.plot(x,f(x))
```

```python
coefficiency1 = np.polyfit(x,noise_y,2)
```

```python
coefficiency1#可以看到对于这个例子，
#使用线性式子生成数值，在拟合时二次项系数几乎为0
```

```python
coefficiency2 = np.polyfit(x,noise_y,3)
coefficiency2#使用三次关系进行拟合三次项系数更小
```

```python
f = np.poly1d(coefficiency2)

p = plt.plot(x, noise_y, 'rx')
plt.plot(x,f(x))
#我们可以对poly1d生成的多项式进行数学运算
print(f)
print(f + 2*f**2)
```
