```python
df = pd.read_csv("http://mlr.cs.umass.edu/ml/machine-learning-databases/event-detection/CalIt2.data", header = None)
```

```python
df.dtypes
#可以看到时间数据竟然是object字符串类型，应该先转换为datetime类型的
```

```python
df.head(5)
```

```python
#使用pd的to_datetime()函数把字符串转换为date_time类型
df[1] = pd.to_datetime(df[1])
df[2] = pd.to_datetime(df[2])
```

```python
df.dtypes
#可以看到都转化为了时间类型了
```

```python
# unique 查看包含了那些独一无二的年 月 日
df[1].dt.year.unique()
```

```python
df[1].dt.month.unique()
```

```python
df[1].dt.day.unique()
```

```python
df[(df[1].dt.day == 1) & (df[1].dt.month == 11)] #可以有多个条件
#!!!!尤其注意括号把条件表达式括好才能避免出现语法问题
#可以单独根据某个时间进行筛选数据
```
