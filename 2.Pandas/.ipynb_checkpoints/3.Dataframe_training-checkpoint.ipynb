```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns 
#seaborn 也是一个画图包
```

```python
df1 = pd.read_csv("train.csv");#其中的NaN是空数据 0/0型的数据
```

```python
df1
```

```python
#查看NaN型数据有多少
df1.isnull().sum().plot('bar')
#Cabin数据中缺失数据最多，其次是年龄数据空的很多，我们需要对这些数据进行处理，才能利用这些数据
```

```python
#查看读入的df的数据类型
df1.dtypes
```

```python
#有计数价值的value_counts
print(df1.Survived.value_counts())
#可以对计数值用plot画出来
df1.Survived.value_counts().plot(kind = 'bar')
```

```python
#可以查看缺失数据多的
df1.isnull().sum().plot('bar')
```

```python
#对于缺失数据多的列可以进行删除操作
df1.drop([1,2],axis = 0) #axis 0是按行删
df1.drop(['Cabin','Name'],axis = 1) #axis 1是按列删
```

```python
#填充缺失数据
df1['Age'] = df1['Age'].fillna(20)
#删除NA多的数据
df1 = df1.drop(['Cabin'],axis = 1)
#
```

```python
#看一看Cabin和 Age列的Na是否都处理了
df1.isnull().sum()
```
