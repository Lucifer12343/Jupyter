```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

meshgrid是常用函数，其功能为
将输入的两个维度的数列阔成一个网格矩阵

```python
x,y = np.meshgrid(np.arange(-10,10,0.05),np.arange(-10,10,0.05))
x
```

```python
y
```

```python
z = np.sqrt(x**2 + y**2)
```

```python
plt.imshow(z)
plt.colorbar()
plt.show()
```

np.where()
where(condition, [x, y])

Return elements, either from `x` or `y`, depending on `condition`.

If only `condition` is given, return ``condition.nonzero()``.

```python
A = np.array([1,2,3,4])
B = np.array([1,8,4,10])
condition = [True,False,False,True]
np.where(condition,A,B)
```

把一个矩阵中小于零的部分都变成0

```python
b = np.random.randn(5,5)
print(b)
np.where(b<0 , 0 , b);
```

数组排序

```python
d = np.random.randn(5)
print(d)
np.sort(d,axis = -1)
```

```python
#np.in1d(a,b) 判断一个一维数组a中的每个元素
#是否在另一个一维数组b中
e = np.array([1,2,3,4,4,5,6])
np.in1d([1,4,4],e)
```

```python
#unique 将1d数组中的每个元素的值取并集
fmat = np.array([1,2,2,'unique'])
print(fmat)
out = np.unique(fmat)
out
```

```python
fmat = np.array([[1,2],[2,'unique']])
print(fmat)
np.unique(fmat,return_index = True)#unique这个函数把二维数组也变成一维的了
```
