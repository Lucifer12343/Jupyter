```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
```

```python
%matplotlib inline
```

```python
#plot(y) 横坐标是1,2,3,4,……
#plot(x, y) 横纵坐标都指定
#plot(x, y, format_string) 横纵坐标,加上绘图格式
x = np.linspace(0, 2 * pi, 51)
plt.plot(np.sin(x))#横坐标
```

```python
plt.plot(x,np.sin(x),x,np.sin(2*x))
```

```python
plt.plot(x,np.sin(x),'r-o') #增加格式标志
```

```python
#使用figure()命令产生新的图像：
t = np.linspace(0, 2*pi, 50)
x = np.sin(t)
y = np.cos(t)
plt.figure()
plt.plot(x)
plt.figure()
plt.plot(y)
```

```python
#使用subplot()在一幅图中画出多个子图
plt.subplot(1, 2, 1)
plt.plot(x)
plt.subplot(1, 2, 2)
plt.plot(y)
```

```python
#label图例,使用legend()显示
plt.plot(t,x,t,y)
plt.legend(['sin','cos'])
```

```python
plt.plot(x,np.sin(x))
plt.xlabel('radians')
# 可以设置字体大小
plt.ylabel('amplitude', fontsize='large')
plt.title('Sin(sin(x))')
```
